## Alternative color image representation

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from skimage.data import astronaut,immunohistochemistry

rgb = astronaut()
r = rgb[:,:,0]
g = rgb[:,:,1]
b = rgb[:,:,2]

plt.gray() # Use grayscale by default on 1-channel images, so you don't have to add cmap=plt.cm.gray everytime

plt.figure()
plt.subplot(2,2,1)
plt.imshow(rgb)
plt.subplot(2,2,2)
plt.imshow(r)
plt.subplot(2,2,3)
plt.imshow(g)
plt.subplot(2,2,4)
plt.imshow(b)
plt.show()
```
