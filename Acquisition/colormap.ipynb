```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
from skimage.data import camera

im = camera()[-1::-2,::2] 
plt.figure(figsize=[5,5])
plt.imshow(im,cmap=cm.gray,interpolation='nearest',origin='lower')
plt.title('gray colormap')

plt.figure(figsize=[5,5])
plt.imshow(im,cmap=cm.jet,interpolation='nearest',origin='lower')
plt.title('jet colormap')

plt.figure(figsize=[5,5])
plt.imshow(im,cmap=cm.hot,interpolation='nearest',origin='lower')
plt.title('hot colormap')

plt.figure(figsize=[5,5])
plt.imshow(im,cmap=cm.prism,interpolation='nearest',origin='lower')
plt.title('prism colormap')


a = np.linspace(0, 1, 256).reshape(1,-1)
a = np.vstack((a,a))

# Get a list of the colormaps in matplotlib.  Ignore the ones that end with
# '_r' because these are simply reversed versions of ones that don't end
# with '_r'
maps = sorted(m for m in cm.datad if not m.endswith("_r"))
nmaps = len(maps) + 1

fig = plt.figure(figsize=(5,10))
fig.subplots_adjust(top=0.99, bottom=0.01, left=0.2, right=0.99)
for i,m in enumerate(maps):
    ax = plt.subplot(nmaps, 1, i+1)
    plt.axis("off")
    plt.imshow(a, aspect='auto', cmap=plt.get_cmap(m), origin='lower')
    pos = list(ax.get_position().bounds)
    fig.text(pos[0] - 0.01, pos[1], m, fontsize=10, horizontalalignment='right')

plt.show()


```
