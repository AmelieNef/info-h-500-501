## Expert System classification

The `text_labels.npy` file contains a pre-labeled version of the text document. The `supervision.json` file contains, for each object, the correct character it represents.

Using this information, and selected features, propose a method to automatically recognize some letters. Count the frequency of each letter in the dataset to determine which characters might be easier to recognize.

*The following example code shows how to use the `text_labels.npy` and `supervision.json` files* :

```python
import json
from skimage.measure import regionprops
from matplotlib.patches import Rectangle

lab = np.load('text_labels.npy')
supervision = json.load(open('super.json','rt'))

plt.figure(figsize=[20,20])
plt.imshow(im,interpolation='nearest',cmap=plt.cm.gray);
ax = plt.gca()

# Simple example that put a box around each "a" character in the image:
objects = regionprops(lab)
for obj in objects:
    str_label = '%s'%obj['label'] # Cast label to string
    if supervision[str_label]=='a':
        bb = obj['bbox']
        centroid = obj['centroid']
        x0 = bb[1]
        y0 = bb[0]
        w = bb[3]-bb[1]+1
        h = bb[2]-bb[0]+1
        ax.add_patch(Rectangle((x0-1,y0-1),w,h,fc='r',ec='r',alpha=.3))
        plt.plot(centroid[1], centroid[0], 'b+')
plt.show()
```
